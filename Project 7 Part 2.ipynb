Chosen Location: Taqueria Mi Pueblito (Mi Puebs)!: (46.07529, -118.31327)

```python
#Now let's use the haversine formula to determine Prof WB location to Mi Puebs!
#Thank you to this stack overflow user for this function: derricw! (https://stackoverflow.com/users/1390022/derricw)
def haversine_np(lon1, lat1, lon2, lat2):
    """
    Calculate the great circle distance between two points
    on the earth (specified in decimal degrees)

    All args must be of equal length.    

    """
    lon1, lat1, lon2, lat2 = map(np.radians, [lon1, lat1, lon2, lat2])

    dlon = lon2 - lon1
    dlat = lat2 - lat1

    a = np.sin(dlat/2.0)**2 + np.cos(lat1) * np.cos(lat2) * np.sin(dlon/2.0)**2

    c = 2 * np.arcsin(np.sqrt(a))
    km = 6367 * c
    return km
```

```python
#Implementing the Formula into WB Locational Data
wb_location["distance_Mi_Puebs"] = haversine_np(wb_location["lon"], wb_location["lat"], -118.31327, 46.07529)
```

```python
#Now let's determine how close professor WB got to Mi-Puebs and the subesquent Timestamp of this moment!
wb_location[wb_location["distance_Mi_Puebs"] == wb_location["distance_Mi_Puebs"].min()]
```

We now know the closest Professor W-B got to Mi Puebs was 0.84 km away from the restuarant at 23:21:58.530000+00:00 on 2022-03-09. 

```python
#What about how far away? 
wb_location[wb_location["distance_Mi_Puebs"] == wb_location["distance_Mi_Puebs"].max()]
```

We now know the furthest away Professor W-B was to Mi Puebs was 1304.884575 km away from the restuarant at 21:34:38.645000+00:00 on 2022-03-12.
